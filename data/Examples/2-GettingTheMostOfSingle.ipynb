## Advanced Single Indexing
In addition to retrieving a single timestep, when calling a DataIndex, if the time is not fully defined, as in that the data is hourly but the time provided is merely day, all timesteps which could fit that criterion are automatically loaded. 
In this call an aggregation function can be provided, which will also be run.

First setting up the ERA5 DataIndex

```python
era5 = edit.data.archive.ERA5(var, level = 'single')
```

There are a variety of ways to now index into this DataIndex. We could specify a particular time.

```python
era5('2021-01-01T01')
```

Retrieving all data on a certain day.

```python
era5('2021-01-01')
```

Providing an aggreagation function to find the 'mean'

```python
era5('2021-01-01', aggregation = 'mean')
```
