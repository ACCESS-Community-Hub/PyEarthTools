# More than a simple time
We have shown how to retrieve a simple single timestep, but how else can we retrieve data?

```python
import edit.data
```

```python
ERA5_index = edit.data.archive.NCI.ERA5('t', level = 'pressure')
ERA5_index
```

`edit` includes a resolution aware datetime implementation, and thus data can be retrieved according to the resolution

```python
time_of_interest = edit.data.EDITDatetime('2022-01-01T00')
time_of_interest
```

```python
ERA5_index[time_of_interest.at_resolution('hour')]
```

```python
ERA5_index[time_of_interest.at_resolution('day')]
```

```python
ERA5_index[time_of_interest.at_resolution('month')]
```

```python
ERA5_index[time_of_interest.at_resolution('year')]
```

## A Series of data
If the data needed does not fit within a resolution bound, it is easy to set your own.

```python
ERA5_index.series(time_of_interest, time_of_interest + edit.data.TimeDelta(10, 'days'), edit.data.TimeDelta(6, 'hours'))
```
